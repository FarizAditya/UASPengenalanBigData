```python
#mengimpor perpustakaan yang di butuhkan.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#disini menggunakan Pandas DataFrame, weather , untuk menampung data cuaca dan memuat file csv ke dalam DataFrame dari url.
#Baris kedua kode mencetak DataFrame dan ini menampilkan data sebagai tabel.
#Melihat tabel Anda dapat melihat bahwa kolom diberi label Tahun , Bulan , Tmax (suhu maksimum), Tmin (suhu minimum), Hujan (curah hujan dalam milimeter) dan Matahari (jam sinar matahari).
weather = pd.read_csv('https://raw.githubusercontent.com/alanjones2/dataviz/master/london2018.csv')
print(weather)
```

```python
#Baris pertama dari kode di atas adalah baris yang berfungsi untuk membuat plot. Itu memanggil metode DataFrame .plot()dari DataFrame weatherdan melewati dua parameter ke fungsi itu, yang pertama adalah nilai untuk y, nilai yang akan diplot secara vertikal, dan yang kedua adalah x, nilai untuk sumbu horizontal.
#Baris kedua kode merujuk ke perpustakaan mathplotlibplt ,. plt.show()adalah fungsi dari perpustakaan itu dan melakukan apa yang Anda harapkan, itu menampilkan grafik.
weather.plot(y='Tmax', x='Month')
plt.show()
```

```python
#dapat melihat bahwa kode ini hampir identik dengan plot pertama tetapi parameter y sekarang berisi nilai untuk suhu maksimum dan minimum ( Tmax dan Tmin ) dalam daftar.
weather.plot(y=['Tmax','Tmin'], x='Month')
```

```python
#Diagram batang
#Untuk menggambar diagram batang dari data yang sama seperti di atas, kami menambahkan teks  .bar ke fungsi plot. Seperti ini:
weather.plot.bar(y='Tmax', x='Month')
```

```python
#Dan jika kita lebih suka bilah horizontal, kita menggunakan versi barh .
weather.plot.barh(y='Tmax', x='Month')
```

```python
#Memplot dua set data (Tmax dan Tmin) dalam diagram batang mirip dengan yang kita lihat di atas dengan diagram garis:
weather.plot.bar(y=['Tmax',"Tmin"], x='Month')
```

```python
#Jika kita ingin membuat satu set grafik terpisah untuk setiap jenis data, kita bisa. 
#Kami mengatur nilai x dan y seperti biasa, tetapi, di samping itu, kami menetapkan subplot parameter sebagai Benar (defaultnya False) dan, 
#jika diinginkan, kami dapat mengatur tata letak seperti yang Anda lihat di bawah ini.
weather.plot.bar(y=['Tmax',"Tmin","Rain","Sun"], x='Month', subplots=True, layout=(2,2))
```

```python
#Kita bisa melakukan ini dengan semua jenis grafik. Inilah hal yang sama tetapi sebagai satu set grafik garis.
weather.plot(y=['Tmax',"Tmin","Rain","Sun"], x='Month', subplots=True, layout=(2,2))
```

```python
#Plot sebar memplot serangkaian poin yang sesuai dengan dua variabel dan memungkinkan kita untuk menentukan apakah ada hubungan di antara mereka. 
#Yang di bawah plot Sun and Rain. Ada 12 titik, satu untuk setiap baris dalam tabel, dan titik-titik memplot nilai Rain pada sumbu vertikal melawan Sun pada horizontal.
weather.plot.scatter(x='Sun', y='Rain')
```
